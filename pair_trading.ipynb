```python
import os
import sys
module_path = os.path.abspath(os.path.join(''))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
import pandas as pd 
import plotly.express as px 
import plotly.graph_objects as go
from auxiliary import tai_pan_dir_to_dataframe_extended
import numpy as np
```

```python
df = tai_pan_dir_to_dataframe_extended(module_path, format="%d.%m.%Y")
```

```python
df = df.apply(lambda x: x.astype(str))
df = df.applymap(lambda x: x.replace(",", "."))
df = df.replace("nan", None).replace("NaN", None)
df = df.apply(lambda x: pd.to_numeric(x))
```

```python
tmp = df["Close"].unstack()
```

```python
# AG : (pref_wkn, ord_wkn)
d = {
    "Sixt AG": ("D23133", "D23132"),
    "Fuchs Petrolub": ("A3E5D6", "A3E5D5"),
    "Volkswagen": ("D66400", "D66403"),
    "Henkel AG": ("C04840", "C04843"),
    "Dr√§gerwerk AG": ("555060", "555063"),
    "Sartorius AG": ("D16560", "D16563")
    }

pref_wkn, ord_wkn = d["Sixt AG"]
tmp = tmp[[pref_wkn, ord_wkn]]
tmp = tmp.dropna()
```

```python
rets = tmp.apply(lambda x: x / x.shift(1) - 1, axis=0)
```

```python
fig = px.scatter(rets.dropna(), x=ord_wkn, y=pref_wkn, width=500, height=500)
fig.show()
```

```python
#rela = tmp
rela = tmp.apply(lambda x: np.log(x), axis=0)
```

```python
rela.loc[:, "diff"] = (rela[ord_wkn] - rela[pref_wkn])
px.line(rela)
```
